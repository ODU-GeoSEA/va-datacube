## **Abstract**

A proof-of-concept for reprojecting a high-resolution external
raster dataset to match the Open Data Cube (ODC) lower-resolution data spatial
grid, as well as another technique for loading and reprojecting ODC data on an
external raster spatial grid. This Google Colab Notebooks uses the **odc_gee**
package to create an ODC-Ready environment in which to test the reprojection
concept using the **datacube** core package.

## **Prepare a Colab ODC-Ready Environemnt**


```python
# Provides Colab with access to your Google Drive to store content and results
from google.colab import drive

drive.mount("/content/drive")

# Grab the Python libraries necessary to run ODC on Google Colab
!wget -nc https://raw.githubusercontent.com/ceos-seo/odc-colab/master/odc_colab.py
from odc_colab import odc_colab_init
from odc_colab import populate_db

odc_colab_init(install_odc_gee=True)
# Populate the ODC data index allowing ODC to access Google Earth Engine Catalog
populate_db()
%pip install xarray[complete] dask dea-tools --upgrade
```

## **Import Required Packages**


```python
import warnings

warnings.filterwarnings("ignore")
import numpy as np
import xarray as xr
from odc_gee import earthengine
from utils.data_cube_utilities.dc_display_map import display_map
from dea_tools.plotting import rgb
from datacube.utils.cog import write_cog
from datacube.testutils.io import rio_slurp_xarray

# Load ODC Configuration
dc = earthengine.Datacube()
```

## **Fill in the Relevant Parameters**


```python
product = "ls8_google"

# Josh drone's sample GeoTIFF
latitude = (38.920814, 38.922512)
longitude = (-104.696715, -104.695000)
time = "2021-01-19"

measurements = ["red", "green", "blue"]
resampling = "nearest"
```

## **View the Area of Interest**


```python
display_map(latitude, longitude)
```

## **Load ODC Data**


```python
query = {
    "product": product,
    "time": time,
    "measurements": measurements,
    "resampling": resampling,
    "group_by": "solar_day",
    "dask_chunks": {"time": "auto", "latitude": "auto", "longitude": "auto"},
}

odc_dataset = dc.load(
    x=longitude,
    y=latitude,
    resolution=(-0.0003, 0.0003),
    **query,
)

odc_dataset
```

## **ODC Data Resolution in Meters**


```python
np.asarray(odc_dataset.geobox.resolution) * 1e5
```

## **Plot ODC Data**


```python
rgb(odc_dataset, bands=["red", "green", "blue"])
```

## **Raster Data Loading**


```python
raster_path = (
    "/content/drive/MyDrive/Colab"
    " Notebooks/NASA/data/drone_0.99cm_epsg4326/SandCreek_Orthomosaic_export_FriMay20155518753439.tif"
)
raster_dataset = xr.open_rasterio(raster_path, chunks="auto").rename(
    {"y": "latitude", "x": "longitude"}
)
raster_dataset = (
    raster_dataset.to_dataset(dim="band")
    .rename({1: "red", 2: "green", 3: "blue"})
    .drop_vars(4)
)
raster_dataset
```

## **Raster Data Resolution in Meters**


```python
np.asarray(raster_dataset.geobox.resolution) * 1e5
```

## **Reproject ODC Data on the Raster Data Spatial Grid**


```python
odc_dataset_on_raster_dataset_grid = dc.load(
    like=raster_dataset.geobox,
    **query,
)

odc_dataset_on_raster_dataset_grid
```

## **ODC Data on Raster Spatial Grid Resolution in Meters**


```python
np.asarray(odc_dataset_on_raster_dataset_grid.geobox.resolution) * 1e5
# rgb(odc_dataset_on_raster_dataset_grid, bands=["red", "green", "blue"])
```

## **ODC Data RGB (COG) Exporting**


```python
# ODC dataset
odc_rgb_da = odc_dataset.isel(time=0)[["red", "green", "blue"]].to_array()
write_cog(
    geo_im=odc_rgb_da,
    fname=(
        "/content/drive/MyDrive/Colab Notebooks/NASA/output/odc_dataset_rgb.tif"
    ),
    use_windowed_writes=True,
    overview_levels=[],
    overview_resampling=resampling,
    overwrite=True,
).compute()
```

## **ODC Data on Raster Grid RGB (COG) Exporting**


```python
# ODC dataset on raster grid
odc_on_raster_grid_rgb_da = odc_dataset_on_raster_dataset_grid.isel(time=0)[
    ["red", "green", "blue"]
].to_array()
write_cog(
    geo_im=odc_on_raster_grid_rgb_da,
    fname=(
        "/content/drive/MyDrive/Colab"
        " Notebooks/NASA/output/odc_dataset_on_raster_grid_rgb.tif"
    ),
    use_windowed_writes=True,
    overview_levels=[],
    overview_resampling=resampling,
    overwrite=True,
).compute()
```

## **Reproject Raster Data on the ODC Data Spatial Grid**


```python
raster_dataset_on_odc_dataset_grid = rio_slurp_xarray(
    raster_path, odc_dataset.geobox, resampling=resampling
)
raster_dataset_on_odc_dataset_grid = (
    raster_dataset_on_odc_dataset_grid.to_dataset(dim="band")
    .rename({0: "red", 1: "green", 2: "blue"})
    .drop_vars(3)
)
raster_dataset_on_odc_dataset_grid
```

## **Plot Rater Dataset on ODC Spatial Grid**


```python
rgb(raster_dataset_on_odc_dataset_grid, bands=["red", "green", "blue"])
```

## **Raster Data on ODC Grid RGB (COG) Exporting**


```python
# Raster dataset on ODC grid
raster_on_odc_grid_rgb_da = raster_dataset_on_odc_dataset_grid[
    ["red", "green", "blue"]
].to_array()
write_cog(
    geo_im=raster_on_odc_grid_rgb_da,
    fname=(
        "/content/drive/MyDrive/Colab"
        " Notebooks/NASA/output/raster_dataset_on_odc_grid_rgb.tif"
    ),
    use_windowed_writes=True,
    overview_levels=[],
    overview_resampling=resampling,
    overwrite=True,
)
```
